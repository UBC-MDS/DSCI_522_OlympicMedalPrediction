```python
import pandas as pd
import numpy as np
from sklearn import tree 
```

```python
data = pd.read_csv("athlete_events.csv")
data = data.query('Season=="Summer"') # Filtered only for Summer season as we are predicting for summer.
data = data.dropna() # Removed all NaN from any row. So we do not have those rows . Is it ok ??
data.head(10)
```

```python
#data.describe()
```

```python
model = tree.DecisionTreeClassifier()
# Adding Features , excluding Year , Name , ID,Games. As Year , Games give the time which will not be needed .
feature_cols = ['Age','Height','Weight']
#feature_cols = ['Age']

X = data.loc[:, feature_cols]
y = data.Medal
                    
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
l = ["paris", "paris", "tokyo", "amsterdam"]
le.fit(l)

print("list(le.classes_)",list(le.classes_))

le.transform(list(le.classes_)) 
#list(le.inverse_transform([0,2, 2, 1,0]))
```

```python
# Fit a model
model.fit(X, y)
```

```python
predictions = model.predict(X)

pred_dict = data.copy()
pred_dict['prediction'] = predictions
pd.DataFrame(pred_dict)
```

```python
model.score(X,y)
```
